```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data_df = pd.read_excel('C:/Users/DELL/Downloads/CP2DataT.xlsx')
x = data_df['xi'].values
y = data_df['yi'].values

# Fit polynomials of degrees 1, 2, and 3
f1 = np.poly1d(np.polyfit(x, y, 1))  # Degree 1
f2 = np.poly1d(np.polyfit(x, y, 2))  # Degree 2
f3 = np.poly1d(np.polyfit(x, y, 3))  # Degree 3

# Calculate the mean of y-values
mean_y = np.mean(y)

# Plotting the data, polynomial fits, and mean line
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Data', color='blue')
plt.plot(x, f1(x), label='Degree 1', color='red')
plt.plot(x, f2(x), label='Degree 2', color='green')
plt.plot(x, f3(x), label='Degree 3', color='orange')
plt.plot(x, np.ones_like(x)*mean_y, label='Mean', linestyle='--', color='black')
plt.xlabel('xi')
plt.ylabel('yi')
plt.title('Data Points, Polynomial Fits, and Mean Line')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def compute_error_metrics(x, y, f, degree):
    mean_y = np.mean(y)
    
    # Calculating the error metrics
    SSres = np.sum((y - f(x))**2)
    SSreg = np.sum((f(x) - mean_y)**2)
    SStot = np.sum((y - mean_y)**2)
    R2 = 1 - SSres/SStot
    
    # Display the results
    print(f"Degree {degree} Polynomial:")
    print(f"SSres: {SSres}")
    print(f"SSreg: {SSreg}")
    print(f"SStot: {SStot}")
    print(f"R^2: {R2}")
    print('-'*40)

# Compute and display error metrics for each polynomial fit
compute_error_metrics(x, y, f1, 1)
compute_error_metrics(x, y, f2, 2)
compute_error_metrics(x, y, f3, 3)
```
